```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
from sb3_contrib import TRPO
from contextlib import contextmanager
# from RL_course.rl_cardiac.tcn_model import TCN_config
from rl_cardiac.tcn_model import TCN_config
from stable_baselines3.common.callbacks import EvalCallback, CheckpointCallback
# from RL_course.rl_cardiac.cardiac_model import CardiacModel_Env
from rl_cardiac.cardiac_model import CardiacModel_Env
from stable_baselines3.common.logger import configure

# Context manager to suppress stdout
@contextmanager
def suppress_stdout():
    original_stdout = sys.stdout
    sys.stdout = open(os.devnull, 'w')
    try:
        yield
    finally:
        sys.stdout.close()
        sys.stdout = original_stdout
```

```python
rat_types = ['healthy_stable', 'hypertension_stable', 'healthy_exercise', 'hypertension_exercise']
SEED = 9
```

```python
rat_type = 'hypertension_stable'
SEED = 9
tcn_model = TCN_config(rat_type)
# env = CardiacModel_Env(tcn_model, rat_type, noise_level=0.1)
env = CardiacModel_Env(tcn_model, rat_type)
env.seed = SEED
env.action_space.seed(SEED)

# TRPO Parameters
policy_kwargs = dict(net_arch=[64])
learning_rate = 0.002
max_kl = 0.01
gamma = 0.99
batch_size = 256
timesteps_per_batch = 2048

# Initialize TRPO
model = TRPO(
    "MlpPolicy",
    env,
    verbose=1,
    learning_rate=learning_rate,
    gamma=gamma,
    batch_size=batch_size,
    policy_kwargs=policy_kwargs,
    target_kl=max_kl,
    n_steps=timesteps_per_batch,
    seed=SEED
)


# Set up logging
log_dir = f"./logs/trpo_experiment/{rat_type}/trial2"
logger = configure(log_dir, ["stdout", "csv", "tensorboard"])
model.set_logger(logger)

# Train or load model
model_path = f"./set_point_trpo_cardiac_models/{rat_type}/trial2.zip"
try:
    model.load(model_path)
except FileNotFoundError:
    print(f"No pretrained model found for {rat_type}, training new model...")
    with suppress_stdout():
        model.learn(total_timesteps=25000)
    model.save(model_path)


results = {}

num_episodes = 10  # Number of episodes
for episode in range(num_episodes):
    print(f"Starting episode {episode + 1}/{num_episodes}")

    # Reset environment for a new episode
    obs = env.reset(seed=SEED)
    if isinstance(obs, tuple):
        obs = obs[0]

    # Episode-specific variables
    time_history = []
    hr_history = []
    map_history = []
    setpoint_hr = []
    setpoint_map = []
    episode_rewards = []

    for t in range(env.ep_length):  # Run for the episode length == 100
        time_history.append(t)

        # Predict action
        action, _ = model.predict(obs, deterministic=True)

        # Take action
        obs, reward, done, truncated, info = env.step(action)
        if isinstance(obs, tuple):
            obs = obs[0]

        # Log HR, MAP, and set-points
        hr, map_ = obs[-4], obs[-3]  # Adjust indices for HR and MAP
        hr_history.append(hr)
        map_history.append(map_)
        setpoint_hr.append(obs[-2])  # HR setpoint
        setpoint_map.append(obs[-1])  # MAP setpoint
        episode_rewards.append(reward)

        # Break if episode ends
        if done or truncated:
            break


    results[episode] = {
    "time": np.array(time_history),
    "hr": np.array(hr_history),
    "map": np.array(map_history),
    "setpoint_hr": np.array(setpoint_hr),
    "setpoint_map": np.array(setpoint_map),
    "episode_rewards": np.array(episode_rewards),
}


# Print results summary
print("Simulation completed.")

# Compute average reward across episodes directly from results
total_rewards_from_results = [sum(results[episode]["episode_rewards"]) for episode in results.keys()]
average_reward_from_results = np.mean(total_rewards_from_results)

print(f"Average reward over {len(results)} episodes: {average_reward_from_results:.2f}")

import matplotlib.pyplot as plt
import numpy as np

# Aggregate results over all episodes
num_episodes = len(results)

# Initialize plots
fig, axes = plt.subplots(2, 1, figsize=(8, 12), sharex=True)
axes = axes.flatten()

# HR and MAP Plots
for episode in range(num_episodes):
    data = results[episode]
    time = data["time"]
    hr = data["hr"]
    map_ = data["map"]
    setpoint_hr = data["setpoint_hr"]
    setpoint_map = data["setpoint_map"]

    # Plot HR
    axes[0].plot(time, hr, label=f"Episode {episode+1}", alpha=0.7)
    axes[0].plot(time, setpoint_hr, "r--", label="HR Set-point" if episode == 0 else "", linewidth=1)

    # Plot MAP
    axes[1].plot(time, map_, label=f"Episode {episode+1}", alpha=0.7)
    axes[1].plot(time, setpoint_map, "r--", label="MAP Set-point" if episode == 0 else "", linewidth=1)

# HR Plot
axes[0].set_title("Heart Rate (HR) over Time")
axes[0].set_ylabel("HR")
axes[0].set_xlabel("Timestep")

# MAP Plot
axes[1].set_title("Mean Arterial Pressure (MAP) over Time")
axes[1].set_ylabel("MAP")
axes[1].set_xlabel("Timestep")

# Adjust layout
plt.tight_layout()
plt.show()

import matplotlib.pyplot as plt
import numpy as np

# Moving average function
def moving_average(data, window_size=50):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# Extract rewards and process for visualization
all_rewards = [results[ep]["episode_rewards"] for ep in range(len(results))]
num_episodes = len(all_rewards)
window_size = 50

# Calculate mean and standard deviation across episodes
mean_rewards = []
std_rewards = []

# Accumulate rewards per iteration
max_length = max(len(rewards) for rewards in all_rewards)
reward_matrix = np.zeros((num_episodes, max_length))

for ep_idx, rewards in enumerate(all_rewards):
    reward_matrix[ep_idx, :len(rewards)] = rewards

# Compute mean and std across episodes
mean_rewards = np.mean(reward_matrix, axis=0)
std_rewards = np.std(reward_matrix, axis=0)

# Smooth mean and standard deviation using a moving average
smoothed_mean = moving_average(mean_rewards, window_size)
smoothed_std = moving_average(std_rewards, window_size)
iterations = np.arange(len(smoothed_mean))

# Plotting
plt.figure(figsize=(10, 6))

# Plot mean reward with shaded region for ± std
plt.plot(iterations, smoothed_mean, label="TRPO Mean Reward", color="blue")

# Labels and Legends
plt.title("TRPO Normalized Reward Values During Training")
plt.xlabel("Iteration")
plt.ylabel("Normalized Reward")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Moving average function
def moving_average(data, window_size=50):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# Extract rewards and process for visualization
all_rewards = [results[ep]["episode_rewards"] for ep in range(len(results))]
num_episodes = len(all_rewards)
window_size = 50

# Calculate mean and standard deviation across episodes
mean_rewards = []
std_rewards = []

# Accumulate rewards per iteration
max_length = max(len(rewards) for rewards in all_rewards)
reward_matrix = np.zeros((num_episodes, max_length))

for ep_idx, rewards in enumerate(all_rewards):
    reward_matrix[ep_idx, :len(rewards)] = rewards

# Compute mean and std across episodes
mean_rewards = np.mean(reward_matrix, axis=0)
std_rewards = np.std(reward_matrix, axis=0)

# Smooth mean and standard deviation using a moving average
smoothed_mean = moving_average(mean_rewards, window_size)
smoothed_std = moving_average(std_rewards, window_size)
iterations = np.arange(len(smoothed_mean))

# Plotting
plt.figure(figsize=(10, 6))

# Plot mean reward with shaded region for ± std
plt.plot(iterations, smoothed_mean, label="TRPO Mean Reward", color="blue")
# plt.fill_between(iterations,
#                  smoothed_mean - smoothed_std,
#                  smoothed_mean + smoothed_std,
#                  color="blue", alpha=0.2, label="± Standard Deviation")

# Labels and Legends
plt.title("TRPO Normalized Reward Values During Training")
plt.xlabel("Iteration")
plt.ylabel("Normalized Reward")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
results = {}

# Loop through each rat_type
for rat_type in rat_types:
    print(f"Simulating for rat_type: {rat_type}")

    tcn_model = TCN_config(rat_type)
    # env = CardiacModel_Env(tcn_model, rat_type, noise_level=0.1)
    env = CardiacModel_Env(tcn_model, rat_type)
    env.seed = SEED
    env.action_space.seed(SEED)

    # Initialize TRPO model
    # policy_kwargs = dict(net_arch=[128, 128])
    # model = TRPO("MlpPolicy", env, verbose=1, learning_rate=0.002, batch_size=16, gamma=0.99, policy_kwargs=policy_kwargs, seed=SEED)

    # policy_kwargs = dict(net_arch=[64])
    # model = TRPO("MlpPolicy", env, verbose=1, learning_rate=0.002, batch_size=16, gamma=0.99, policy_kwargs=policy_kwargs, seed=SEED)

    # TRPO Parameters
    policy_kwargs = dict(net_arch=[64])
    learning_rate = 0.002
    max_kl = 0.01
    gamma = 0.99
    batch_size = 256
    timesteps_per_batch = 2048

    # Initialize TRPO
    model = TRPO(
        "MlpPolicy",
        env,
        verbose=1,
        learning_rate=learning_rate,
        gamma=gamma,
        batch_size=batch_size,
        policy_kwargs=policy_kwargs,
        target_kl=max_kl,
        n_steps=timesteps_per_batch,
        seed=SEED
    )


    # Set up logging
    log_dir = f"./logs/trpo_experiment/{rat_type}"
    logger = configure(log_dir, ["stdout", "csv", "tensorboard"])
    model.set_logger(logger)

    # Train or load model
    model_path = f"./set_point_trpo_cardiac_models/{rat_type}.zip"
    try:
        model.load(model_path)
    except FileNotFoundError:
        print(f"No pretrained model found for {rat_type}, training new model...")
        with suppress_stdout():
            model.learn(total_timesteps=25000)
        model.save(model_path)

    # Logging variables
    time_history = []
    hr_history = []  # Heart Rate
    map_history = []  # Mean Arterial Pressure
    setpoint_hr = []  # Target HR
    setpoint_map = []  # Target MAP

    # Evaluation
    obs = env.reset(seed=SEED)
    if isinstance(obs, tuple):
        obs = obs[0]

    for t in range(200):  # Simulate for 200 iterations
        time_history.append(t)

        # Predict action
        action, _ = model.predict(obs, deterministic=True)

        # Take action
        obs, _, done, truncated, _ = env.step(action)
        if isinstance(obs, tuple):
            obs = obs[0]

        # Log HR, MAP, and set-points
        hr, map = obs[0], obs[1]  # Adjust indices based on observation format
        hr_history.append(hr)
        map_history.append(map)
        setpoint_hr.append(env.setpoints[0])
        setpoint_map.append(env.setpoints[1])

        if done or truncated:
            break

    # Save results
    results[rat_type] = {
        "time": np.array(time_history),
        "hr": np.array(hr_history),
        "map": np.array(map_history),
        "setpoint_hr": np.array(setpoint_hr),
        "setpoint_map": np.array(setpoint_map),
    }
    print(f"Results for {rat_type}:\n {results[rat_type]}\n")
```
